```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(np.arange(5),np.arange(5)) #x轴的范围（0-4），y轴的范围（0-4）
plt.show()
```

```python
#xy随机生成0到1符合正态分布的500个点
x = np.random.normal(0,1,500)
y = np.random.normal(0,1,500)

#s=50设定点的大小，c=‘b’表示颜色为蓝色，alpha=0.5表示透明度为50%
plt.scatter(x,y,s=50,c='b',alpha=0.5)

#限制图像的大小
plt.xlim((-2,2))
plt.ylim((-2,2))

#去除xy的尺距（和数字）
plt.xticks(())
plt.yticks(())

plt.show()
```
