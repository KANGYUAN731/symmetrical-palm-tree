### 假设2010年1月1日开始，每月第一个交易日买入1手股票，每年最后一个交易日卖出股票求收益
 - 时间节点：2010-2020
 -  一手股票=100股股票
 - 买：一个完整的年需要买入12*100=1200
 - 卖：一个完整的年需要卖出12*100=1200
 - 单价：使用开盘价

```python
new_df = df['2010-01':'2020-02']
new_df
```

```python
#买股票：每个月的第一个交易日对应的行数据（捕获开盘价格）
#每个月第一个交易日=每个月第一行的数据
#根据月份从原始数据中提取每月第一个交易日的数据
df_monthly = new_df.resample('M').first()
df_monthly
```

 - 数据中出现了01-31日字样，但其实数据没有错误只是date的索引出现了问题

```python
new_df.head(2)
```

```python
#买入股票的花费的总金额
cost = df_monthly['open'].sum()*100
cost
```

```python
#卖出股票到手的钱
#每年最后一个交易日
df_yearly = new_df.resample('A').last()
df_yearly
```

```python
resv = df_yearly['open'].sum()*1200
resv
```

```python
#计算总收益
resv - cost
```
